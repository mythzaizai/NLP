```python
import json
from opencc import OpenCC
import jieba
import re
from gensim.models import word2vec
```

```python
r = '[）\（\：\·\，\。\“ \”\?\？\」\「\……\、\《\》\；\)\(]'
file = ['AA','AB','AC','AD','AE','AF','AG','AH','AI','AJ','AK','AL']
cc = OpenCC('s2t')

def open_wiki():
    result =[]
    for a in file[0:4]:
        for i in range(100):
            for line in open('wiki_zh/{}/wiki_{}'.format(a,str(i).zfill(2))):
                data = json.loads(line)
                value = re.sub(r,'',data['text'])
                result.append(value)

    for a in file[5:8]:
        for i in range(100):
            for line in open('wiki_zh/{}/wiki_{}'.format(a,str(i).zfill(2))): 
                data = json.loads(line)
                value = re.sub(r,'',data['text'])
                result.append(value)

    for a in file[9:11]:
        for i in range(100):
            for line in open('wiki_zh/{}/wiki_{}'.format(a,str(i).zfill(2))):
                data = json.loads(line)
                value = re.sub(r,'',data['text'])
                result.append(value)

    for i in range(74):
        for line in open('wiki_zh/{}/wiki_{}'.format('AM',str(i).zfill(2))):
            data = json.loads(line)
            value = re.sub(r,'',data['text'])
            result.append(value)
            
    return result
    
result = open_wiki()

print(len(result))
```

```python
def chine(result):
    value = []
    for line in result:
        a = cc.convert(line)
        value.append(a)
    return value

value=chine(result)
```

```python
def get_stopword_list():       
    with open('wiki_zh/cn_stopwords.txt') as f: 
        stopword_list = [word.strip('\n') for word in f.readlines()]
    return stopword_list
```

```python
def macktxt(value):
    stop = get_stopword_list()
    f = open('test1.txt','w')

    print(len(value))

    jieba.set_dictionary('/usr/local/lib/python3.9/site-packages/jieba/dict.txt')
    jieba.load_userdict("txt2.txt")  

    for line in value :  
        data = jieba.lcut(line,cut_all=False)
        for val in data:
            if val not in stop:
                f.write(val+' ')
        # f.write('\n')
    f.close

macktxt(value)
```

```python
def word2():
    train_data = word2vec.LineSentence('test.txt')
    # seed = 666
    sg = 0
    # window_size = 10
    vector_size = 300
    # min_count = 1
    workers = 4
    epochs = 5
    # batch_words = 10000
    model = word2vec.Word2Vec(
        train_data,
        # min_count=min_count,
        vector_size=vector_size,
        workers=workers,
        epochs=epochs,
        # window=window_size,
        sg=sg,
        # seed=seed,
        # batch_words=batch_words
    )
    model.save('word2vec.model')

```

```python
def test(string):
    model = word2vec.Word2Vec.load('word2vec.model')
    # print(model.wv['生物'].shape)

    for item in model.wv.most_similar(string,topn=20):
        print(item)
```

```python
word2()
print('訓練成功！')
test('地球')
```
